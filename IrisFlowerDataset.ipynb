```python
# Optional styling
sns.set(style="whitegrid")
```

```python
plt.figure()
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm')
plt.title("Sepal Length vs Width")
plt.show()
```

```python
df.hist(figsize=(10, 8))
plt.suptitle("Histograms of Features")
plt.show()
```

```python
sns.histplot(df['SepalLengthCm'], kde=True)
plt.title("Distribution of feature1")
plt.show()
```

```python
sns.boxplot(x=df['SepalLengthCm'])
plt.title("Box Plot of feature1")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.xticks(rotation=45)
plt.title("Box Plot for All Numeric Features")
plt.show()
```
